```python
from pathlib import Path
import pandas as pd
from sklearn import preprocessing
from sklearn.model_selection import train_test_split
from sklearn.metrics import accuracy_score
from sklearn.neighbors import NearestNeighbors, KNeighborsClassifier
import matplotlib.pylab as plt

iris=pd.read_csv('iris.csv')
iris.columns=['sepallength','sepalwidth','petallength','petalwidth','class']
iris['Number']=iris.index+1
trainData, validData= train_test_split(iris,test_size=0.3, random_state=26)
newIris=pd.DataFrame([{'sepallength':5,'sepalwidth':3,'petallength':1.5,'petalwidth':0.3}])
predictors= ['sepallength','sepalwidth','petallength','petalwidth']

scaler=preprocessing.StandardScaler()
scaler.fit(trainData[['sepallength','sepalwidth','petallength','petalwidth']])
irisNorm=pd.concat([pd.DataFrame(scaler.transform(iris[['sepallength','sepalwidth','petallength','petalwidth']]),
        columns=['zsepallength','zsepalwidth','zpetallength','zpetalwidth']),iris[['sepallength','sepalwidth','petallength','petalwidth','class']]],axis=1)
trainNorm=irisNorm.iloc[trainData.index]
validNorm=irisNorm.iloc[validData.index]
newIrisNorm=pd.DataFrame(scaler.transform(newIris),columns=['zsepallength','zsepalwidth','zpetallength','zpetalwidth'])

knn=NearestNeighbors(n_neighbors=3)
knn.fit(trainNorm.iloc[:,0:4])
distances,indices=knn.kneighbors(newIrisNorm)
trainNorm.iloc[indices[0],:]

trainX=trainNorm[['zsepallength','zsepalwidth','zpetallength','zpetalwidth']]
trainy=trainNorm['class']
validX=validNorm[['zsepallength','zsepalwidth','zpetallength','zpetalwidth']]
validy=validNorm['class']
results=[]
for k in range(1,15):
    knn=KNeighborsClassifier(n_neighbors=k).fit(trainX,trainy)
    results.append({
        'k':k,
        'accuracy':accuracy_score(validy,knn.predict(validX))
    })
results=pd.DataFrame(results)
print(results)
```
